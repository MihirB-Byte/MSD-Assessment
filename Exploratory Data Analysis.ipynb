```python
#Cleaning the data 
```

```python
injury_df = injury_df.drop_duplicates()
injury_df
```

```python
missing_injury =injury_df.isna().sum()
missing_injury
```

```python
injury_df = injury_df.dropna(subset=['Injury'])
injury_df
```

```python
injury_df = injury_df.drop(index=injury_df[injury_df['Injury'] == 'Unassigned'].index)
injury_df
```

```python
# Convert the Incident Date column to datetime and derived new columns year, month and day from it.

event_df['Incident Date'] = pd.to_datetime(event_df['Incident Date'], format='%d/%m/%Y')
event_df['year'] = event_df['Incident Date'].dt.year
event_df['month'] = event_df['Incident Date'].dt.month
event_df['day'] = event_df['Incident Date'].dt.day

event_df.info()
print(event_df['Incident Date'])
```

```python
#Now that we've cleansed the data we can start analysing the data.
```

```python
#Distribution of demographics by gender

gender_counts = demographic_df["Gender"].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Distribution of demographics by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#Distribution of demographics by ethnicity

ethnicity_counts = demographic_df["Ethnicity"].value_counts()
plt.bar(ethnicity_counts.index, ethnicity_counts.values)
plt.title("Distribution of demographics by Ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#Distribution of demographics by Age Band

age_counts = demographic_df["Age Band"].value_counts()
plt.bar(age_counts.index, age_counts.values)
plt.title("Distribution of demographics by Age Band")
plt.xlabel("Age Band")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#Distribution of demographics by Salary Range

sal_counts = demographic_df["Salary Range"].value_counts()
plt.bar(sal_counts.index, sal_counts.values)
plt.title("Distribution of demographics by Salary Range")
plt.xlabel("Salary Range")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#Regions with the highest incident and lowest incident rates

#To achieve this we need to mmerge the data frames to get the information we need. 

#We need to update the column name of 'Person Employee Name' to 'Employee Name in the events df to match the column name in demographics df

event_df = event_df.rename(columns={'Person Employee Name': 'Employee Name'})
event_df.head()

```

```python
# merge the injury and event dataframes on the Incident ID column

injury_event_df = pd.merge(injury_df, event_df, on='Incident ID')
injury_event_df.head()
```

```python
# merge the injury_event and demographic dataframes on the Employee Name column

final_df = pd.merge(injury_event_df, demographic_df, on='Employee Name')
final_df.head()
```

```python
# rate = number of incidents per employee
# calculate incident rates by region

incident_rates_by_region = final_df.groupby('Site 2 Name')['Incident ID'].count() / final_df.groupby('Site 2 Name')['Employee Name'].nunique()

# sort regions by incident rates in descending order

sorted_regions = incident_rates_by_region.sort_values(ascending=False)
```

```python
# create a bar chart of incident rates by region

plt.bar(sorted_regions.index, sorted_regions.values)
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Incident Rate')
plt.title('Regions with highest and lowest incident rates')
plt.show()
```

```python
# total incidents by region
incidents_by_region = final_df.groupby("Site 2 Name")["Incident ID"].count()

# sort by count in descending order

incidents_by_region = incidents_by_region.sort_values(ascending=False)

# plot the results

incidents_by_region.plot(kind="bar", figsize=(10, 6))
plt.title("Total Incidents by Region")
plt.xlabel("Region")
plt.ylabel("Total Incidents")
plt.show()
```

```python
# Summary on the top 5 type of incidents

# get the top 5 incident types

top_incidents = final_df["Injury"].value_counts().head(5)

# plot the results

top_incidents.plot(kind="bar", figsize=(10, 6))
plt.title("Top 5 Incident Types")
plt.xlabel("Incident Type")
plt.ylabel("Total Incidents")
plt.show()
```

```python
# Top 5 common body parts injured

# get the top 5 injured body parts

top_body_parts = final_df["Body Part"].value_counts().head(5)

# plot the results

top_body_parts.plot(kind="bar", figsize=(10, 6))
plt.title("Top 5 Injured Body Parts")
plt.xlabel("Body Part")
plt.ylabel("Total Injuries")
plt.show()
```

```python
# Top 5 Position Titles with most injuries

# Group by Position Type and count the number of injuries

position_counts = final_df.groupby('Position Title')['Injury'].count()

# Sort in descending order and get the top 5 positions

top_5_positions = position_counts.sort_values(ascending=False).head(5)

# Plot a bar chart of the top 5 positions
plt.bar(top_5_positions.index, top_5_positions.values)
plt.title('Top 5 Position Types with the Most Injuries')
plt.xlabel('Position Type')
plt.xticks(rotation=90)
plt.ylabel('Number of Injuries')
plt.show()
```

```python
# calculate injury count by salary range

# Count the number of injury incidents by salary range

injury_count_by_salary = final_df.groupby('Salary Range')['Injury'].count().reset_index()

# Sort the values by count in descending order

injury_count_by_salary = injury_count_by_salary.sort_values(by='Injury', ascending=False)

# Create a bar plot

plt.bar(injury_count_by_salary['Salary Range'], injury_count_by_salary['Injury'])
plt.xlabel('Salary Range')
plt.ylabel('Number of Injury Incidents')
plt.title('Number of Injury Incidents by Salary Range')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Finding the timeline of the dataset

# Fetching Earliest date and Latest date from the data set

earliest_date = final_df['Incident Date'].min()
latest_date = final_df['Incident Date'].max()

print('Earliest date:', earliest_date)
print('Latest date:', latest_date)
```

```python
# plotting the frequency of incidents

final_df['date'] = pd.to_datetime(final_df['Incident Date'])

final_df.set_index('date', inplace=True)

# Resample the data by week and count the number of incidents per week
weekly_counts = final_df.resample('W')['Incident ID'].count()

# Plot the time series
weekly_counts.plot()
plt.xlabel('Week')
plt.ylabel('Number of incidents')
plt.title('Timeline of incidents')
plt.show()
```

```python
# Monthly frequency

# Resample the data by month and count the number of incidents per month
monthly_counts = final_df.resample('M')['Incident ID'].count()

# Plot the time series
monthly_counts.plot()
plt.xlabel('Month')
plt.ylabel('Number of incidents')
plt.title('Timeline of incidents')
plt.show()
```

```python
# Plotting the locations of injuries

# create a new dataframe with injury counts by site

injury_by_site = final_df['Site 1 Name'].value_counts()

# plot the distribution of injuries by site
plt.figure(figsize=(8,6))
sns.barplot(x=injury_by_site.index, y=injury_by_site.values)
plt.title('Injury Distribution by Site')
plt.xlabel('Site Name')
plt.ylabel('Injury Count')
plt.xticks(rotation=90)
plt.show()
```

```python
final_df['Site 1 Name'].unique()
```

```python
# type of events captured at sites


# filter for a specific site name
site_name = "Client Service Delivery"
site_df = final_df[final_df["Site 1 Name"] == site_name]

# check if site_df is empty before proceeding
if site_df.empty:
    print("No data available for site:", site_name)
else:
    # group by injury type and get the count of each type of injury
    injury_counts = site_df.groupby("Event Type Name")["Incident ID"].count().sort_values(ascending=False)

    # plot the results as a bar chart
    injury_counts.plot(kind="bar")
    plt.title("Types of Injuries at {}".format(site_name))
    plt.xlabel("Injury Type")
    plt.ylabel("Count")
    plt.show()
```

```python
# Visualising the incident status

# group by incident status and get the count of each status

incident_status_counts = final_df.groupby("Incident Status")["Incident ID"].count()

# create a bar to show the distribution of incident statuses

incident_status_counts.plot(kind="bar")
plt.title("Distribution of Incident Statuses")
plt.xlabel("Incident Statuses")
plt.ylabel("Number of Cases")
plt.show()
```

```python
# Visualising the Potential consequences of the incidents

# group by Potential consequences and get the count of each consequences

potential_consequence_counts = final_df.groupby("Potential Consequence")["Incident ID"].count()

# create a bar to show the distribution of Potential Consequence

potential_consequence_counts.plot(kind="bar")
plt.title("Distribution of Potential Consequence")
plt.xlabel("Potential Consequences")
plt.ylabel("Number of Cases")
plt.show()
```

```python
# Visualising the Treatments of Injuries

# group by Treatments of Injuries and get the count of each treatment

potential_consequence_counts = final_df.groupby("Treatment of Injury")["Incident ID"].count()

# create a bar to show the distribution of incident statuses

potential_consequence_counts.plot(kind="bar")
plt.title("Distribution of Treatment of Injury")
plt.xlabel("Treatments")
plt.ylabel("Number of Cases")
plt.show()
```

```python
# Visualising the Nature of Injuries

# group by Nature of Injuries and get the count of each Nature

injury_nature_counts = final_df.groupby("Name")["Incident ID"].count()

# create a bar to show the distribution of incident statuses

injury_nature_counts.plot(kind="bar")
plt.title("Distribution of Nature of Injury")
plt.xlabel("Names")
plt.ylabel("Number of Cases")
plt.show()
```

```python
# Visualising the Ergonimic Nature of Injuries

# group by Nature of Injuries and get the count of each Nature for Ergonomic type of injuries  

ergonomic_df = final_df[final_df["Name"] == "Ergonomic"]

subcategory_counts = ergonomic_df.groupby("Sub Category")["Incident ID"].count().sort_values(ascending=False)


# create a bar to show the distribution of incident statuses

subcategory_counts.plot(kind="bar")
plt.title("Distribution of Nature of Injury for Ergonomic Type")
plt.xlabel("Sub Category")
plt.ylabel("Number of Incidents")
plt.show()

```
